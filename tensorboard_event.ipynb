```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator

# Replace this with the directory containing your event files.
event_file_path = "EXP_ONLY_CORPUS/chv_rus_1M_Corpus_model/tensorboard/events.out.tfevents.1683985826.cn-004.107878.0"

# Initialize an event accumulator.
event_acc = EventAccumulator(event_file_path)

# Reload the event file.
event_acc.Reload()

# List all the scalar tags that are available.
print(event_acc.Tags()["scalars"])

# Get the values for a specific scalar tag.
# Replace 'tag_name' with the name of the tag you want to extract.
tag_name = 'bleu-val'
values = event_acc.Scalars(tag_name)

# 'values' is a list of ScalarEvent(wall_time, step, value) tuples.
# You can extract the step values and the corresponding scalar values like this:
steps = [scalar_event.step for scalar_event in values]
scalars = [scalar_event.value for scalar_event in values]
```

```python
import matplotlib.pyplot as plt

steps = [scalar_event.step for scalar_event in values]
scalars = [scalar_event.value for scalar_event in values]

plt.figure(figsize=(10,5))
plt.plot(steps, scalars)
plt.xlabel('Steps')
plt.ylabel('val-bleu')
plt.title('val-bleu over Steps')
plt.grid(True)
plt.show()
```
